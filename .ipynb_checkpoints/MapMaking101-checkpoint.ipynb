```python
# Custom tiles

from branca.element import Figure
fig=Figure(width=1200,height=500)

centre = [51.529163, -0.109970]

m1=folium.Map(width=600, 
              height=500,
              location= centre, 
              zoom_start=14,
              tiles='cartodbpositron'
         )
m2 = folium.Map(width=600, 
              height=500,
              location= centre, 
              zoom_start=14,
             tiles='OpenStreetMap'
         )

fig.add_child(m1)
fig.add_child(m2)
```
